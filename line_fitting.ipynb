## Line Fitting

```python
# Numerical arrays.
import numpy as np

#Plotting 
import matplotlib.pyplot as plt
```

A Straight line in two dimensions can always be written as:

$y = mx + c$

# Data

https://numpy.org/doc/stable/reference/generated/numpy.polyfit.html


```python
# Create x values
x = np.linspace(0.0, 10.0, 31)

# Create y values
y = 3.0 * x
```

```python
# Create scatterplot

fig, ax = plt.subplots()

# Plot the data
ax.scatter(x, y)
```

```python
# use polyfit to fit a line to the data
m, c = np.polyfit(x, y, 1)
```

```python
# Create scatterplot

fig, ax = plt.subplots()

# Plot the data
ax.scatter(x, y)

ax.plot(x, m *x + c, color='red')
```

https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html

```python
# Create x values
x = np.linspace(0.0, 10.0, 31)

# Create y values
y = 3.0 * x + np.random.normal(0.0, 0.3, size=x.size) 
# this is noise 
```

```python
# Create scatterplot

fig, ax = plt.subplots()

# Plot the data
ax.scatter(x, y)
```

```python
# use polyfit to fit a line to the data
m, c = np.polyfit(x, y, 1)

# show
m, c
```

```python
# Create scatterplot

fig, ax = plt.subplots()

# Plot the data
ax.scatter(x, y)

ax.plot(x, m *x + c, color='red')
```

```python
# Create x values
x = np.linspace(0.0, 10.0, 31)

# Create y values
y = 3.0 * x + 10.0 + np.random.normal(0.0, 1.0, size=x.size) 

# use polyfit to fit a line to the data
m, c = np.polyfit(x, y, 1)

# show
m, c

# Create scatterplot

fig, ax = plt.subplots()

# Plot the data
ax.scatter(x, y)

ax.plot(x, m *x + c, color='red')

ax.set_title(f'y = {m:.2f}x + {c:.2f}')
```
