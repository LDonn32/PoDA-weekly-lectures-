```python
# Plots.
import matplotlib.pyplot as plt

# Numerical arrays.

import numpy as np
```

```python
# X values.
x = np.linspace(0.0, 10.0, 101)

# Show.
x
```

```python
# Y values 
y = x**2

# double * raises to the power of two
# numpy goes into each array and raises to power of two

#show
y
```

```python
# Plot x vs y
plt.plot (x, y, label='$y = x^2$'); 

# semi colan gets rid of the output [<matplotlib.lines.Line2D at 0x7505a48cabd0
# dollar signs read it in as latex code, more technial and mathmathecial way
# this is a graph to represent x squared

plt.legend();
plt.xlabel('x')
plt.ylabel('$x^2$');

plt.title(' A Lovely Plot');
```
