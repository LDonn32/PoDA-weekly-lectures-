# Plotting functions

```python
# Plots.
import matplotlib.pyplot as plt

# Numerical arrays.

import numpy as np
```

```python
# X values.
x = np.linspace(0.0, 10.0, 101)

# Show.
x
```

```python
# Y values 
y = x**2

# double * raises to the power of two
# numpy goes into each array and raises to power of two

#show
y
```

```python
# Plot x vs y
plt.plot (x, y, label='$y = x^2$'); 

# semi colan gets rid of the output [<matplotlib.lines.Line2D at 0x7505a48cabd0
# dollar signs read it in as latex code, more technial and mathmathecial way
# this is a graph to represent x squared

# Add a legend
plt.legend();

# Add labels
plt.xlabel('x')
plt.ylabel('$x^2$');

# Add title
plt.title(' A Lovely Plot');
```

## comparing plots

```python
# Domain of x values 
x = np.linspace(0.0, 10.0, 101)

# x sqaured
y1 = x**100 + x**2 + x + 1


# 2 to the power of x
y2 = 2**x



```

```python
# plot x vs y1
plt.plot(x, y1, label= '$ x^2$') # for $ latex code to work have to change double * to valid latex code = ^

# plot x vs y2
plt.plot(x, y2, label='$ 2^x$')

# Add a legend.
plt.legend()

# Add labels.
plt.xlabel('x')
plt.ylabel('y')

# Add a title.
plt.title('Two Lovely Plots');
```

## Integers

```python
# Domains of x values
x = np.arange(1, 101, 1)

# show.
x
```

```python
# The type
x.dtype
```

```python
# Domains of x values
# https://numpy.org/doc/stable/reference/generated/numpy.linspace.html
x = np.linspace(0, 10, 1001, dtype=int)

# show.
x
```

```python
# The type
x.dtype
```

```python
# x cubed plus 2 x squared plus 1.
y = x**3 + 2 * x**2 + 1
```

```python
# Type of y
y.dtype
```

```python
# Plot x vs y.
plt.plot(x, y, '.', label='$x^3 + 2x^2 + 1$') 

# '.' shows data as dots rather than a line. more appropriate for ints

# Add a legend.
plt.legend()

# Add labels.

plt.xlabel('x')
plt.ylabel('y')

# add a title
plt.title('Plotting ints.');
```
