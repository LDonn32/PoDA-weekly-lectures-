```python
# Numerical arrays.
import numpy as np

# Plotting
import  matplotlib.pyplot as plt 
```

# Simple Linear Regression

$y = mx + c$

$y = 2x + 1$

```python
# An array of x values.
x = np.linspace(0.0, 10.0, 21)

# An array of y values.
y = 3.0 * x + np.random.normal(0.0, 1.0, x.size)
```

```python
# Create fig and axis
fig, ax = plt.subplots()

# Plot the data. 
ax.scatter(x, y, marker='.', label ='$y = 2x + 1$')

# Add legend.
ax.legend()
```

```python
# Polyfit
np.polyfit(x, y, 1)
```

# Calculation from Wikipedia

https://en.wikipedia.org/wiki/Simple_linear_regression#Formulation_and_computation

```python
# X mean
xbar = np.mean(x)
# Y mean
ybar = np.mean(y)
```

```python
# Numeriator 
num = np.sum((x - xbar) * (y - ybar))

# Denominator.
den = np.sum((x - xbar) ** 2)

# Slope
m = num / den

# Show.
m
```

```python
# Intercept.
c = ybar - (m * xbar)

# Show.
c
```

$y = mx$

```python
# Bunch of x values.
x = np.linspace(0.0, 10.0, 21)

# Corresponding y values.
y = 3.0 * x + np.random.normal(0.0, 1.0, x.size)
```

```python
# Plot.
fig, ax = plt.subplots()

# Scatter.
ax.scatter(x, y, marker='.', label='$y = 3x$')

# Add the line that we know should be the best fit.
ax.plot(x, 3.0 * x, 'r', label='True best fit line')

# Legend.
ax.legend();
```

```python
# Fit.
np.sum(x * y) / np.sum(x * x)
```

# Approximating




```python
# X values.
x = np.linspace(0.0, 10.0, 21)

# Y values.
y = 3.0 * x + np.random.normal(0.0, 1.0, x.size)

# Cost function.
def cost(m):
    return np.sum((m * x - y) ** 2)

# Show cost for m = 3.0.
cost(3.0)
```

```python
# A range of m values.
m = np.linspace(-3.0, 9.0, 101)

# Calculate the cost function for each m.
costs = np.array([cost(mi) for mi in m])

# Plot the cost function.
fig, ax = plt.subplots()
ax.plot(m, costs);
```

```python
# Optimize m using a dirty hack.
guess = 1.0

# Learning rate.
alpha = 0.01

# Keep going.
keep_going = True

# No iterations.
iterations = 0

while keep_going:
    iterations += 1
    if cost(guess + alpha) < cost(guess):
        guess += alpha
    elif cost(guess - alpha) < cost(guess):
        guess -= alpha
    else:
        keep_going = False

# Show.
guess, iterations
```

```python
# Fit.
np.sum(x * y) / np.sum(x * x)
```
