```python
# Plotting
import matplotlib.pyplot as plt

# data manipulation
import numpy as np

# Machine learning
from sklearn import datasets
```

```python
# https://scikit-learn.org/stable/datasets/toy_dataset.html
diabetes = datasets.load_diabetes(scaled=False)

# Show keys.
diabetes.keys()
```

```python
#  feature names

diabetes.feature_names
```

```python
# Data
diabetes.data
```

```python
# BMI
bmi = diabetes.data[:, 2]

# show first ten 
bmi[:10]
```

```python
# blood pressure
bp = diabetes.data[:, 3] # use the [:, 3] :, and olum number three to show all the datat as a column. If we just use [3]it will only give us the row

# Show
bp[:10]
```

```python
# Create a figure and an axis
fig, ax = plt.subplots()

# Scatter plot
ax.scatter(bmi,bp)

# Labels.
ax.set_xlabel('BMI')
ax.set_ylabel('BloodPressure')
```

```python
# Create a figure and an axis
fig, ax = plt.subplots()

# Scatter plot
ax.scatter(bmi,bp, marker = 'v') # this marker makes changes the data plots style

# Labels.
ax.set_xlabel('BMI')
ax.set_ylabel('BloodPressure')
```

```python
# sex
sex = diabetes.data[:, 1]

# show first ten
sex[:10]
```

```python
# Unique sex values
np.unique(sex)
```

```python
# Create a figure and an axis
fig, ax = plt.subplots()

# Scatter plot
# change colour
ax.scatter(bmi,bp, c = sex) 

# Labels.
ax.set_xlabel('BMI')
ax.set_ylabel('BloodPressure')
```

##  Scale

```python
# Age
age = diabetes.data[:,0]

#  Show first 10
age[:10]
```

```python
# Create a figure and an axis
fig, ax = plt.subplots()

# Scatter plot
# change colour
ax.scatter(bmi,bp, c = sex, s=age, marker='.') 

# Labels.
ax.set_xlabel('BMI')
ax.set_ylabel('BloodPressure')
```
